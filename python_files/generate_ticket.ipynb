```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
from docx import Document
from docx.shared import Inches

def generate_receipt(money_given, date, No_land, date_payment):
    # --- Rutas ---
    base_de_datos = r"C:\\Users\\azy51\\OneDrive\Documentos\\Comprobante las jarrilleras\\Las jarrilleras.xlsx"
    imagen_recibo = r"C:\\Users\\azy51\\OneDrive\Documentos\\Comprobante las jarrilleras\\Captura de pantalla 2025-10-19 094707.png"
    
    # --- Leer Excel ---
    df = pd.read_excel(base_de_datos, sheet_name="Registro")
    
    # --- Buscar fila del cliente y lote ---
    fila = df[ (df["LOTE"] == No_land)]
    if fila.empty:
        print("No se encontró ese cliente y lote en la base de datos.")
        return
    fila_index = fila.index[0]

    name = df.loc[fila_index, "CLIENTE"]
    
    # --- Calcular nuevo pago restante ---
    pago_restante_nuevo = df.loc[fila_index, "PAGO RESTANTE"] - money_given
    
    # --- Actualizar Excel ---
    df.loc[fila_index, "CANTIDAD RECIBIDA"] = money_given
    df.loc[fila_index, "PAGO RESTANTE"] = pago_restante_nuevo
    df.loc[fila_index, "FECHA"] = date
    df.to_excel(base_de_datos, sheet_name="Registro", index=False)
    

    
    # --- Mostrar recibo ---
    img = mpimg.imread(imagen_recibo)
    plt.figure(figsize=(4,4))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    
    print(f"Fecha: {date}\n")
    print(f"Nombre del cliente: {name}\n")
    print(f"Lote {No_land} en el paraje La Jarrillera\n")
    print(f"Pago recibido: {money_given}\n")
    print(f"Pago restante: ${pago_restante_nuevo}\n")
    print("NOTA: Este recibo respalda su dinero y su inversión. Favor de cuidarlo y presentarlo para cualquier aclaración.")
    print(f"Su proxima fecha de pago es: {date_payment}")


    doc = Document()
    doc.add_picture(imagen_recibo, width=Inches(3.5))
    doc.add_paragraph()
    doc.add_paragraph(f"Fecha: {date}")
    doc.add_paragraph(f"Nombre del cliente: {name}")
    doc.add_paragraph(f"Lote {No_land} en el paraje La Jarrillera")
    doc.add_paragraph(f"Pago recibido: {money_given}")
    doc.add_paragraph(f"Pago restante: ${pago_restante_nuevo}")
    doc.add_paragraph("NOTA: Este recibo respalda su dinero y su inversión. "
                  "Favor de cuidarlo y presentarlo para cualquier aclaración.")
    doc.add_paragraph(f"Su proxima fecha de pago es: {date_payment}")

    output_path = r"C:\Users\azy51\OneDrive\Documentos\Comprobante las jarrilleras\recibo.docx"
    doc.save(output_path)



```

```python
generate_receipt(2000,"15/08/2020",46, "15/09/2025")
```
